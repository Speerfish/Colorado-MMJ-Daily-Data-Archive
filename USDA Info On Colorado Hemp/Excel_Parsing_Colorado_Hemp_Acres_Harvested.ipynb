```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 1: Load Excel File

```python
# Load the Excel file into a Pandas DataFrame
file_path = 'Hemp_Acres_Harvested.xlsx'

# Read the file into a dataframe
df = pd.read_excel(file_path, usecols=['Year', 'Data Item', 'Value'])

# Display the first few rows of the dataframe to check the data
df.head(15)
```

## Step 2: Clean and Prepare the Data

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values (if needed)
df.dropna(inplace=True)

# Ensure the 'Year' column is an integer and 'Value' is a float
df['Year'] = df['Year'].astype(int)
df['Value'] = df['Value'].astype(float)

# Display the cleaned data
df.head(15)
```

## Step 3: Plot Bar Graphs by Year

```python
# Get the unique years
years = df['Year'].unique()

# Loop through each year and create individual bar graphs
for year in years:
    # Filter data for the current year
    year_data = df[df['Year'] == year]
    
    # Create a new figure for each year
    plt.figure(figsize=(12, 8))
    
    # Plot the bar graph for the current year
    plt.bar(year_data['Data Item'], year_data['Value'], color='green')
    
    # Add labels and title for the current year's plot
    plt.xlabel('Data Item')
    plt.ylabel('Acres (Value)')
    plt.title(f'Hemp Acres Harvested in {year}')
    
    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Show the plot
    plt.tight_layout()
    plt.show()

plt.savefig('hemp_acres_harvested.png', dpi=300)
```
